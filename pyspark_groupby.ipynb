```python
import findspark
findspark.init()
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder\
    .appName("NNS7")\
    .getOrCreate()
```

```python
df = spark.read.json("./data/raw/drivers.json")
df.show()
```

```python
waldf = spark.read.option('header','true').csv('./data/walmart/walmart_stock.csv')
```

```python
waldf.show()
```

```python
from pyspark.sql.functions import year, month,max

waldf = waldf.withColumn("year", year(waldf["Date"])).withColumn("Month", month(waldf["Date"]))
waldf.show()
```

```python
maxdf = waldf.groupBy("year") \
     .agg(max("Volume"))

```
